```python
# Import dependencies
import feyn
import pandas as pd
import numpy as np
# from sklearn.datasets import load_diabetes
```

```python
I_11_19_data = pd.read_csv("I_11_19_data.csv")
```

```python
X = I_11_19_data.loc[1:1000, ['x1', 'x2', 'x3', 'y1', 'y2', 'y3']]
X = X.reset_index(drop=True)
# y = x1*x4 + x2*x5 + x3*x6
y = I_11_19_data.loc[1:1000, ['I.11.19_data$Y']]
y = y.squeeze(axis=0)
y = y.iloc[:,0]
y = y.rename("x")
y = y.reset_index(drop=True)

# Create a DataFrame
dataset = pd.DataFrame({
    'x1': X['x1'],
    'x2': X['x2'],
    'x3': X['x3'],
    'y1': X['y1'],
    'y2': X['y2'],
    'y3': X['y3'],
    'y': y
})
```

```python
# Train/test split
train, test = feyn.tools.split(dataset, ratio=[0.8, 0.2], random_state=42)

# Instantiate a QLattice
ql = feyn.QLattice(random_seed=42)

# Sample and fit models
models = ql.auto_run(
    data=train,
    output_name='y'
)

# Best model
best = models[0]

best.plot(train, test)
```
